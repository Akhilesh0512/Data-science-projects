```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in df.columns[:-1]:  # Excluding the 'Class' column.
    plt.figure(figsize=(10, 4))
    
    # Plotting the distribution of non-fraud transactions
    sns.histplot(df[df['Class'] == 0][column], bins=50, color='blue', alpha=0.5, label='Not Fraud')
    
    # Plotting the distribution of fraud transactions
    sns.histplot(df[df['Class'] == 1][column], bins=50, color='red', alpha=1, label='Fraud')
    plt.title(f'Histogram for {column}')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming this is how you've generated your data
np.random.seed(0)
total_data_points = 284807  # Total number of data points

scater_plot = pd.DataFrame({
    'Time': np.linspace(0, 172800, total_data_points),
    'Amount': np.random.exponential(scale=200, size=total_data_points),
    'Class': np.random.choice([0, 1], size=total_data_points, p=[0.99, 0.01])
})
bins = np.linspace(0, max(scater_plot['Amount']), 30)

# Now proceed with the histogram plotting as before
plt.figure(figsize=(20, 12))
plt.hist(scater_plot[scater_plot['Class'] == 0]['Amount'], bins=bins, alpha=0.5, label='Not Fraud', color='blue')
plt.hist(scater_plot[scater_plot['Class'] == 1]['Amount'], bins=bins, alpha=1, label='Fraud', color='red')
plt.legend()
plt.title('Transaction Amounts by Class')
plt.xlabel('Amount ($)')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):   
    
    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    # print
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}")
    print(f"num_cols: {len(num_cols)}")
    print(f"cat_but_car: {len(cat_but_car)}")
    print(f"num_but_cat: {len(num_but_cat)}")

    return cat_cols, num_cols, cat_but_car


cat_cols, num_cols, cat_but_car = grab_col_names(df)
```

```python
def num_summary(dataframe, numerical_col, plot=True):
    quantiles = [0, 0.01, 0.10, 0.25, 0.50, 0.75, 0.95, 0.99, 1]
    print(dataframe[numerical_col].describe(quantiles).T)

    if plot:
        dataframe[numerical_col].hist(bins=20)
        plt.xlabel(numerical_col)
        plt.title(numerical_col)
        plt.show(block=True)
for col in num_cols:
    num_summary(df, col, plot=False)
```
